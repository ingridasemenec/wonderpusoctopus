```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
```

```python
# Creates folder called images 
path_im = Path.cwd()/'images'
if path_im.exists() == False:
    path_im.mkdir()
```

```python
# Creates folders with features names under images folder
data_var = ['fe','no3','o2','si','ph','chl']
for x in data_var:
    if (path_im/x).exists() == False:
        (path_im/x).mkdir()
```

```python
# Load the pisces dataset the xarray form from local 
XR = xr.load_dataset('PISCES.nc')
XR = XR.mean(dim = 'depth')
```

```python
XR
```

```python
# Fill in the folders you have created before with time sorted pics
for i in np.arange(XR['time'].size):
    for var in data_var:
        plt.rcParams["figure.dpi"] = 72
        fig, axes = plt.subplots(ncols=1)
        XR.sel(time = XR['time'][i])[var].plot(cmap='gray',add_colorbar = False)
        plt.axis('off')
        plt.title(None)
        fig.savefig(path_im/var/(str(XR['time'][i].values).partition('T')[0].replace("-", "_")+'.png')
                , bbox_inches='tight', pad_inches=0)
        plt.close()
```

```python
# Creates a gif from the feature you choose 
import imageio.v2 as imageio
ims = [imageio.imread(f) for f in list(sorted((path_im/'chl').glob('*.png')))]
if (path_im/'gifs').exists() == False:
    (path_im/'gifs').mkdir()
imageio.mimwrite(path_im/'gifs'/'chl.gif', ims, fps=5)
```
